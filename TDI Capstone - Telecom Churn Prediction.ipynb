```python
fig, axs = plt.subplots(3,3,figsize = (20,20))

axs[0,0].hist(df[df['churn']==1]['rev_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[0,0].hist(df[df['churn']==0]['rev_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[0,0].set_title('mean monthly revenue')
axs[0,0].legend(['Churn', 'Non-Churn'])

axs[0,1].hist(df[df['churn']==1]['totmrc_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[0,1].hist(df[df['churn']==0]['totmrc_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[0,1].set_title('mean total monthly recurring charge')
axs[0,1].legend(['Churn', 'Non-Churn'])

axs[0,2].hist(df[df['churn']==1]['change_mou'],range = (0,200), bins = 50, alpha = 0.5)
axs[0,2].hist(df[df['churn']==0]['change_mou'],range = (0,200), bins = 50, alpha = 0.5)
axs[0,2].set_title('% change in monthly min. of use vs prev. 3 month ave.')
axs[0,2].legend(['Churn', 'Non-Churn'])

axs[1,0].hist(df[df['churn']==1]['change_rev'],range = (0,200), bins = 50, alpha = 0.5)
axs[1,0].hist(df[df['churn']==0]['change_rev'],range = (0,200), bins = 50, alpha = 0.5)
axs[1,0].set_title('% change in monthly revenue vs prev. 3 month ave.')
axs[1,0].legend(['Churn', 'Non-Churn'])

axs[1,1].hist(df[df['churn']==1]['blck_vce_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[1,1].hist(df[df['churn']==0]['blck_vce_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[1,1].set_title('mean number of blocked (failed) voice calls')
axs[1,1].legend(['Churn', 'Non-Churn'])

axs[1,2].hist(df[df['churn']==1]['custcare_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[1,2].hist(df[df['churn']==0]['custcare_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[1,2].set_title('mean number of customer care calls')
axs[1,2].legend(['Churn', 'Non-Churn'])

axs[2,0].hist(df[df['churn']==1]['ccrndmou_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[2,0].hist(df[df['churn']==0]['ccrndmou_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[2,0].set_title('mean rounded minutes of use of customer care calls')
axs[2,0].legend(['Churn', 'Non-Churn'])

axs[2,1].hist(df[df['churn']==1]['complete_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[2,1].hist(df[df['churn']==0]['complete_Mean'],range = (0,200), bins = 50, alpha = 0.5)
axs[2,1].set_title('mean number of completed calls')
axs[2,1].legend(['Churn', 'Non-Churn'])

axs[2,2].hist(df[df['churn']==1]['months'],range = (0,200), bins = 50, alpha = 0.5)
axs[2,2].hist(df[df['churn']==0]['months'],range = (0,200), bins = 50, alpha = 0.5)
axs[2,2].set_title('Total number of months in service')
axs[2,2].legend(['Churn', 'Non-Churn'])

plt.show()
```

### Takeaway 0
As seen in the histograms above, features behave mostly similarly in terms of frequency, which makes coming up with new derivative features a must. 

```python
labels = ['Minute Usage Change','Completed Calls', 'Overage Revenue', 'Minutes Customer Care', 'Handset Price', 'Tot. Time of Service']
churn_values = [df[df['churn']==1]['change_mou'].mean(),
                df[df['churn']==1]['mou_cvce_Mean'].mean(), 
                df[df['churn']==1]['ovrrev_Mean'].mean(), 
                df[df['churn']==1]['ccrndmou_Mean'].mean(), 
                df[df['churn']==1]['hnd_price'].mean(),
                df[df['churn']==1]['months'].mean(),
               ]
nonchurn_values = [df[df['churn']==0]['change_mou'].mean(), 
                   df[df['churn']==0]['mou_cvce_Mean'].mean(), 
                   df[df['churn']==0]['ovrrev_Mean'].mean(), 
                   df[df['churn']==0]['ccrndmou_Mean'].mean(), 
                   df[df['churn']==0]['hnd_price'].mean(),
                   df[df['churn']==0]['months'].mean(),
               ]
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize = (15,5))
rects1 = ax.bar(x - width/2, churn_values, width, label='Churn')
rects2 = ax.bar(x + width/2, nonchurn_values, width, label='Non-Churn')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Normalized Percent')
# ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# ax.bar_label(rects1, padding=3)
# ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

### Takeaway 1-2
The features in the bar plot above, except the last feature, have been chosen such that the difference between the mean values are the most among other features. For example, first figure tells that the first signal a subscriber gives before the churn is to talk less. This data is averaged over the last 3 months. Similarly, churn customers completed calls are significantly lower than non-churns.
### Takeaway 3
Second figure tells that those who churn are paying more for overage than those who don't on average. This could be a sign that the customer's current plan do not fit his/her needs
### Takeaway 4
One could expect for churn customers to talk to customer care services more. However, data suggests the other way around, which implies that those who touch base with the customer care are more likely to solve their problems and stay.
### Takeaway 5
Customers with more expensive handsets are more likely to stay, which is worth further exploration.
### Takeaway 6
Average amoun of time that churn and nonchurn customers have been served has no significant difference, which suggests that the amount of time passed with a company do not have any effect on churn decision, i.e. low customer loyalty. 

--------------------------------
These takeaways clearly show that the data in hand is worth working on and it is possible to acquire significant accuracies with the help of some feature engineering.
