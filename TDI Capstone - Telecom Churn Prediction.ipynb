```python
# Returns the appropriate size of plots grid depending on the number of plots
def plot_grid(f):
    h = int(np.sqrt(f))
    k = int(f/h)
    if k*h < f:
        if k > h:
            h+=1
        else:
            k +=1
    return k,h
```

```python
# Create plots in a grid for initial data exploration purposes
plot_features = ['rev_Mean','totmrc_Mean','change_mou','change_rev','blck_vce_Mean','custcare_Mean','ccrndmou_Mean','complete_Mean','months']
x_plots, y_plots = plot_grid(len(plot_features))
fig, axs = plt.subplots(x_plots,y_plots,figsize = (20,20))
feature_index = 0

for handle_x in range(x_plots):
    for handle_y in range(y_plots):
        feature = plot_features[feature_index]
        axs[handle_x,handle_y].hist(df[df['churn']==1][feature], range=(0,200), bins=50, alpha=0.5)
        axs[handle_x,handle_y].hist(df[df['churn']==0][feature], range=(0,200), bins=50, alpha=0.5)
        axs[handle_x,handle_y].set_title(columns_dict[feature])
        axs[handle_x,handle_y].legend(['Churn', 'Non-Churn'])
        feature_index += 1

plt.show()
```

### Takeaway 0
As seen in the histograms above, features behave mostly similarly in terms of frequency, which makes coming up with new derivative features a must. 

```python
features = ['mou_cvce_Mean','ovrrev_Mean','ccrndmou_Mean','hnd_price','months']#'change_mou','adjrev','adjmou','adjqty','avgrev','avgmou','avgqty','avg3mou','avg3qty','avg3rev','avg6mou','avg6qty','avg6rev']
labels = []
for f in features:
    labels.append(columns_dict[f])
    
churn_values = []
nonchurn_values = []
for f in features:
    churn_values.append(df[df['churn']==1][f].mean())
    nonchurn_values.append(df[df['churn']==0][f].mean())
    
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize = (15,10))
rects1 = ax.bar(x - width/2, churn_values, width, label='Churn')
rects2 = ax.bar(x + width/2, nonchurn_values, width, label='Non-Churn')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Average Values')
# ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

# ax.bar_label(rects1, padding=3)
# ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

### Takeaway 1-2
The features in the bar plot above, except the last feature, have been chosen such that the difference between the mean values are the most among other features. For example, first figure tells that the first signal a subscriber gives before the churn is to talk less. This data is averaged over the last 3 months. Similarly, churn customers completed calls are significantly lower than non-churns.
### Takeaway 3
Second figure tells that those who churn are paying more for overage than those who don't on average. This could be a sign that the customer's current plan do not fit his/her needs
### Takeaway 4
One could expect for churn customers to talk to customer care services more. However, data suggests the other way around, which implies that those who touch base with the customer care are more likely to solve their problems and stay.
### Takeaway 5
Customers with more expensive handsets are more likely to stay, which is worth further exploration.
### Takeaway 6
Average amount of time that churn and nonchurn customers have been served has no significant difference, which suggests that the amount of time passed with a company do not have any effect on churn decision, i.e. low customer loyalty.

## Initial ML to understand the base
Let us give it a try and train a simple model without doing any feature engineering (feature selection/aggregation etc).

```python
df.isnull().sum().plot()
```

```python
# DROP the columns that has more than 5% null values
df_init = df.drop(df.columns[df.isnull().sum()>len(df)*0.05], axis = 1)
```

```python
df_init.isnull().sum().max()
```

```python
len(df_init.columns)
```

```python
# DROP the rows that has *any* null value
df_init.dropna(inplace = True)
```

```python
len(df_init)
```

```python
df_init.isnull().sum().sum()
```

```python
df_cat = df_init
```

```python
# EXTRACT the numerical columns
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
df_init = df_init.select_dtypes(include=numerics)
```

```python
# Unique value frequency percentages for categorical columns
for column in df_init.columns:
    if len(df_cat[column].unique()) < 3:
        print(columns_dict[column], '-->', df_cat[column].unique())
        print(df_cat[column].value_counts(normalize=True) * 100)
        print('\n')
```

<div class="alert alert-block alert-info">
<b>Correlation Analysis:</b> Correlation analysis between numerical-categorical and categorical-categorical is tricky, requires further attention
</div>

```python
fig = plt.figure(figsize=(30,30))
sns.heatmap(df_init.corr())
plt.show()
```

```python
# CREATE THE CORRELATION MATRIX and filter with more than 90% correlation
df_init_corr = (np.abs(df_init.corr())>0.9) & (np.abs(df_init.corr())<1)
```

```python
# CREATE the set that has names of highly correlated columns
corr_set = set()
for column in df_init_corr:
    if len(df_init_corr[column][df_init_corr[column]==True]) >0:
#         print(columns_dict[column], '  ',df_init_corr[column][df_init_corr[column]==True].index)
        for item in df_init_corr[column][df_init_corr[column]==True].index:
            corr_set.add(item)
#     corr_dict[column] = df_init_corr[column][df_init_corr[column]==True].index
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

### First create a model using all the numerical columns without removing anything

```python
# remove the Cutomer_ID column as it is a random number assigned to every customer
y = df_init['churn']
X = df_init.drop(['churn', 'Customer_ID'], axis=1)
```

```python
# standardize the values to help models converge easily
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X))
```

```python
X.head()
```

### Begin trying LOGISTIC REGRESSION

```python
lr=LogisticRegression(solver="lbfgs", max_iter=1000, random_state=2)

y_pred = cross_val_predict(lr,X, y, cv=5)
```

```python
print('Accuracy Score:  ', accuracy_score(y,y_pred).round(2))
print(confusion_matrix(y, y_pred))
print('\n',classification_report(y,y_pred))
```

```python
# Half of the set CHURNS, dataset is BALANCED
round((1 - sum(df_init['churn']==1)/len(df_init)),3)
```

#### Let's remove correlated columns and try again LOGISTIC REGRESSION

```python
df_init_corr = df_init.drop(corr_set, axis = 1)
```

```python
df_init_corr.columns
```

```python
y_corr = df_init_corr['churn']
X_corr = df_init_corr.drop(['churn', 'Customer_ID'], axis=1)
scaler = StandardScaler()
X_corr = pd.DataFrame(scaler.fit_transform(X_corr))
```

```python
lr_corr=LogisticRegression(solver="lbfgs", max_iter=1000, random_state=2)
y_pred_corr = cross_val_predict(lr_corr,X_corr, y_corr, cv=5)

```

```python
print('Accuracy Score:  ', accuracy_score(y_corr,y_pred_corr).round(2))
print(confusion_matrix(y_corr, y_pred_corr))
print('\n',classification_report(y_corr,y_pred_corr))
```

#### We used only the numerical columns ready and no categorical columns. We dealt with the null values roughly. Logistic regression gives only a ***7%*** improvement. Not a good job actually! 

Now let a tree classifier decide which columns are important

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier(n_estimators=20)
model.fit(X_corr, y_corr)
print(model.feature_importances_)
```

```python
feature_importances = pd.Series(model.feature_importances_)
feature_importances.plot()
```

```python
feature_importances.describe()
```

```python
features_cust = feature_importances[feature_importances >=0.02].index
```

```python
lr=LogisticRegression(solver="lbfgs", max_iter=1000, random_state=2)

scores = cross_val_score(lr,X[features_cust], y, cv=5)
scores.mean()
```

## LET'S START WORKING ON THE CATEGORICAL COLUMNS

```python
df_cat = df_cat.select_dtypes(exclude=numerics)
df_cat['churn'] = df_init['churn']
```

```python
df_cat.head()
```

```python
# Unique value frequency percentages for categorical columns
for column in df_cat.columns:
    print(columns_dict[column], '-->', df_cat[column].unique())
    print(df_cat[column].value_counts(normalize=True) * 100)
    print('\n')
```

```python
# Create plots in a grid for initial data exploration purposes
plot_features = list(df_cat.columns)
plot_features.remove('churn')
# x_plots, y_plots = plot_grid(len(plot_features))
fig, axs = plt.subplots(len(plot_features),1,figsize = (20,50))
fig_index = 0
tot = len(df_cat)
width = 0.35
for feature in plot_features:
    x = np.arange(len(df_cat[feature].unique()))
    vals = df_cat[feature].value_counts().index
    churn_values = []
    nonchurn_values = []
    
    for unique_val in vals:
        
        if unique_val in df_cat[feature][df_cat['churn']==1].value_counts(normalize=True).index:
            churn_value = df_cat[feature][df_cat['churn']==1].value_counts(normalize=True).loc[unique_val]*100
        else:
            churn_value = 0
            
        if unique_val in df_cat[feature][df_cat['churn']==0].value_counts(normalize=True).index:
            nonchurn_value = df_cat[feature][df_cat['churn']==0].value_counts(normalize=True).loc[unique_val]*100
        else:
            nonchurn_value = 0
        churn_values.append(churn_value)
        nonchurn_values.append(nonchurn_value)
            
    rects1 = axs[fig_index].bar(x - width/2, churn_values, width, label='Churn')
    rects2 = axs[fig_index].bar(x + width/2, nonchurn_values, width, label='Non-Churn')
    axs[fig_index].set_xticks(x)
    axs[fig_index].set_xticklabels(vals)
    axs[fig_index].legend()
    axs[fig_index].set_title(columns_dict[feature])
    fig_index += 1

plt.show()
```
